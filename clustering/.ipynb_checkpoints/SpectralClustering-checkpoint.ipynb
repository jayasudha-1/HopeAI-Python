```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled= sc.fit_transform(X)
```

```python
from sklearn.cluster import SpectralClustering
clusmodel = SpectralClustering(n_clusters=5, assign_labels='kmeans',
                               eigen_solver='arpack', random_state=0)
label = clusmodel.fit_predict(X_scaled)
```

```python
label
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=label
```

```python
supervised
```

```python
supervised.to_csv("SpectralClustering.csv",index=False)
```
